```python
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob
import nltk
from nltk.corpus import stopwords
import readability

idx = 0
for link in links:
    url = link
    response = requests.get(url)

    soup = BeautifulSoup(response.content, "html.parser")

    text = soup.get_text()

    blob = TextBlob(text)

    polarity_score = blob.sentiment.polarity
    subjectivity_score = blob.sentiment.subjectivity

    filtered_words = [word for word in blob.words.lower() if word not in stop_words]

    pos_score = 0
    neg_score = 0
    for word in filtered_words:
        if TextBlob(word).sentiment.polarity > 0:
            pos_score += 1
        elif TextBlob(word).sentiment.polarity < 0:
            neg_score += 1

    fog_index = readability.getmeasures(text)['readability grades']['GunningFogIndex']

    df['POSITIVE SCORE'][idx] = pos_score
    df['NEGATIVE SCORE'][idx] = neg_score
    df['POLARITY SCORE'][idx] = polarity_score
    df['SUBJECTIVITY SCORE'][idx] = subjectivity_score
    df['FOG INDEX'][idx] = fog_index

    idx +=1


```
